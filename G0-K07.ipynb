```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import sympy as sp
import numpy as np
import scipy.optimize as opt
from sympy.utilities.lambdify import lambdify
import scipy.integrate as integrate
from numpy.lib.scimath import sqrt
sp.init_printing()
```

```python
sk = sp.Symbol('k', real = True)
sx = sp.Symbol('x', real = True)
sh = sp.Symbol('hbar', real = True)
sm = sp.Symbol('m', real = True)
sw = sp.Symbol('omega', real = True)
st = sp.Symbol('t', real = True)
sE = sp.Symbol('E', real = True)
sw = sp.Symbol('w', real = True)
sP = sp.Function('Psi')
```

```python
def aplus(psi):
    return 1 / sp.sqrt(2 * sm * sh * sw) * (-sh * psi.diff(sx) + sm * sw * sx * psib)
```

```python
psi0 = (sm * sw / (sp.pi * sh))**sp.Rational(1, 4) * sp.exp(-sm * sw * sx**2 / sh)
psi0
```

```python
psi0f = lambdify(sx, psi0.subs({sm: 1, sw: 1, sh: 1}))
```

```python
xs = np.linspace(-5, 5, 1000)
```

```python
plt.plot(xs, psi0f(xs))
```

```python
psi1 = aplus(psi0) / sp.sqrt(1)
psi1
```

```python
psi1f = lambdify(sx, psi1.subs({sm: 1, sw: 1, sh: 1}))
plt.plot(xs, psi1f(xs))
plt.plot(xs, 0.5 * xs * xs)
plt.hlines(3/2, -5, 5)
plt.ylim(-1, 2)
```

```python
psi2 = aplus(psi1) / sp.sqrt(2)
psi2.simplify()
```

```python
psi2f = lambdify(sx, psi2.subs({sm: 1, sw: 1, sh: 1}))
plt.plot(xs, psi2f(xs))
```
